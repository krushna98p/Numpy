# copy vs View in numpy

#    copy 
copy own tne data.<br>the copy of an array is a new array.<br>the change made in the copy data does not reflect in the original array

# view
the view does not own data.<br>
a view of the original array.<br>any changes made to the view will affect the original array.and any change made to the original array will affect the view

```python
# copy
arr=np.array([1,2,3,4,5,6])
co=arr.copy()
print('original:',arr)
print('copy:',co)

arr[2]=10
print('original after change:',arr)
print('copy:',co)             
```

```python
# view
varr=np.array([1,2,3,4,5,6])
vi=varr.view()
print('original:',varr)
print('view:',vi)

varr[2]=10
print('original after change:',varr)
print('view:',vi)             
```

```python
j1=np.array([1,2,3,4,5])
j2=np.array([7,8,9])
jo=np.concatenate((j1,j2))
print(jo)
```

```python
j3=np.array([[1,2,],[3,4]])
j4=np.array([[5,7],[8,9]])
joo=np.concatenate((j3,j4),axis=0)
print(joo)
```

```python
jos=np.stack((j3,j4),axis=1)
print(jos)
```

```python
jor=np.hstack((j3,j4))
print(jor)
print()

joc=np.stack((j3,j4))
print(joc)
print()
joh=np.stack((j3,j4))
print(joh)
```
