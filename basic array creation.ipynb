# array creation

```python
import numpy as np
list1=[1,2,3,4,5,6]
arr=np.array(list1)
print(arr)
```

# 2D array & 3D array

```python
list2=[list1]
arr2=np.array(list2)
list3=[list2]
arr3=np.array(list3)
print(arr2)
print(arr3)
```

# multiple dimention array

```python
arr4=np.array(list1,ndmin=10)
print(arr4)
```

#  check array dimention

```python
print(np.ndim(arr4))
```

# create array of zeross

```python
arr5=np.zeros((4,4))
print(arr5)
```

# create array of onces

```python
arr6=np.ones((4,5))
print(arr6)
```

# empty array

```python
arr7=np.empty((4,5))
print(arr7)                     #it copy privious if same dimention passed
```

# range array

```python
arr8=np.arange(4)
print(arr8)
```

# identity matrix or only diagonal element array

```python
arr9=np.eye(4,5)
print(arr9)
```

# linspace funtion makes same spaced array

```python
arr10=np.linspace(0,20, num=5)
print(arr10)
```

# random number array creation

# rand() function
generate values between 0 to 1

```python
# generate values between 0 to 1
ran1=np.random.rand(4)
print(ran1)
```

```python
ran2=np.random.rand(3,3)
print(ran2)
```

# randn() function
generate number near to zero

```python
ran3=np.random.randn(4)
print(ran3)
```

```python
ran4=np.random.randn(3,3)
print(ran4)
```

# ranf() function
generate value between 0.0 to 1.0 but it excludes 1

```python
ran5=np.random.ranf(4)
print(ran5)
```

# randint(min,max,total value) function
it generates randon integer between given range

```python
ran6=np.random.randint(5,20,5)
print(ran6)
```
