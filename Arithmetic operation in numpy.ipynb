# Arithmetic operation

#    operation                     function   
## 1. A+B-------------------------np.add(A,B)
## 2. A-B-------------------------np.substarct(A,B)
## 3. A*B-------------------------np.multiply(A,B)
## 4. A/B-------------------------np.divide(A,B)
## 5. A%B-------------------------np.mod(A,B)
## 6. A**B------------------------np.power(A,B)
## 7. 1/A-------------------------np.reciprocal(A)


```python
import numpy as np 
arr1=np.array([1,2,3,4,5])
arr2=np.array([3,4,5,6,7])
```

### 1. Add

```python
add1=arr1+arr2
print(add1)
```

```python
add2=np.add(arr1,arr2)
print(add2)
```

### 2. substract

```python
sub1=arr1-arr2
print(sub1)
```

```python
sub2=np.subtract(arr1,arr2)
print(sub2)
```

### 3. Multiply

```python
mul1=arr1*arr2
print(mul1)
```

```python
mul2=np.multiply(arr1,arr2)
print(mul2)
```

### 4. Divide

```python
div1 =arr1/arr2
print(div1)
```

```python
div2=np.divide(arr1,arr2)
print(div2)
```

### 5. Modulo

```python
mod1=arr1%arr2
print(mod1)
```

```python
mod2=np.mod(arr1,arr2)
print(mod2)
```

### 6. Power

```python
pov1=arr1**arr2
print(pov1)
```

```python
pov2=np.power(arr1,arr2)
print(pov2)
```

### 7. Inverse or resiprocal

```python
res1=1/arr1
print(res1)
```

```python
res2=np.reciprocal(arr1)
print(res2)
```

# Arithmetic function

## 1. np.min(x)-----------------------  find minimum element from array
## 2. np.max(x)-----------------------  find maximum element from array 
## 3. np.argmin/max(x)----------------  find address of min or max element
## 4. np.sqrt(x)----------------------  find squreroot of each value of array
## 5. np.sin(x)-----------------------  find sin value of each element
## 6. np.cos(x)-----------------------  find cos value of each element
## 7. np.cumsum(x)--------------------  adds vext value in element expect index 0

```python
x=np.array([1,2,3,4,5,6,7])

print('minimum element:',np.min(x))
print('maximum element:',np.max(x))
print('mimimmum element address:',np.argmin(x))
print('squreroot :',np.sqrt(x))
print('sin value',np.sin(x))
print('cos value:',np.cos(x))
print('cumsum:',np.cumsum(x))
```
