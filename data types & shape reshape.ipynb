# data types in numpy array

# for check data type use dtype function

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
print(arr1.dtype)
```

```python
arr2=np.array([1.0,2.3])
print(arr2.dtype)
```

# datatype changing methods
# 1. direct method

```python
arr3=np.array([1,2,3,4,5],dtype=float)
print(arr3.dtype)
```

# 2. shortcut method

```python
arr4=np.array([1,2,3,4,5],dtype="f")
print(arr4.dtype)
```

# 3. as function

```python
arr5=np.array([1,2,3,4,5])
arr5=np.float64(arr5)
print(arr5.dtype)
```

# 4. astype() function

```python
arr6=np.array([1,2,3,4,5])
arr6=arr5.astype(complex)
print(arr6,arr6.dtype)
```

# Shape of array

```python
sh1=np.array([[1,2,3,4,5]])
print('dimention :',np.ndim(sh1))
print('shape:',sh1.shape)
# it means 1 row andd 5 columns
```

# Reshape of array

```python
sh2=np.array([[1,2,3,4,5,6]])
sh2=sh2.reshape(3,2)
print(sh2)
print('dimention :',np.ndim(sh2))
print('shape:',sh2.shape)
```

```python
sh3=sh2.reshape(-1)
print(sh3)
print('dimention :',np.ndim(sh3))
print('shape:',sh3.shape)
```
