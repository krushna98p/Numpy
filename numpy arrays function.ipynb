# Search Array
Search an array for certain value, and return the indexes that get a match.<br>
for seach we use where key word<br>
you can also search by passing aritmetic operation <br>
its return index of element

```python
arr=np.array([1,4,2,3,4,5,644,4,7,8])
x=np.where(arr==4)
print(x)
```

# search sorted array
which performs a binary search in the array, and returns the index where the specified value would be inserted to maintain the search order.<br>
it uses np.searchsorted(var,element,side) function we can pass full ins in it and we can aslo pass where its search.

```python
arr1=np.array([1,2,4,6,7,9])
s=np.searchsorted(arr1,[3,5,8])  #side is left by default
s
```

# Flatten & Ravel
both conver 2d array in singal diamention array<Br>
it has row major order anc column major order <br>
order{C,F,A,K}<br>
C---- means to flatten in row major(c style) order<br>
F---- means to flatten in column major (fortran style)order<br>
A---- means to flatten in column major if 'a' is Fortran *contiguous* in memory, row major order otherwise.<br>
K---- means to flatten 'a' in the order the elements occur in memory 


<br><br>
by default it is 'C'

```python
print('Flatten:',re.flatten(order='F'))
print('Ravel:',np.ravel(re,order='F'))
```
