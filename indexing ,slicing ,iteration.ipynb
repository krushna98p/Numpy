# Indexing
## 1D
## 2D
## 3D

```python
import numpy as np
arr1=np.array([1,2,3,4,5])
print('array:',arr1)
print('dia:',arr1.ndim)
print('geting 3:',arr1[2])
```

```python
arr2=np.array([[1,2],[4,5]])
print('array:\n',arr2)
print('dia:',arr2.ndim)
print('geting 4:',arr2[1,0])
```

```python
arr3=np.array([[[1,2,3],[4,5,6]]])
print('array:\n',arr3)
print('dia:',arr3.ndim)
print('geting 3:',arr3[0,0,2])
```

# Slicing
## variable[start:stop:step]
its also itrate n-1 index <br>
1.from start  to stop= variable[ : stop]<br>
2. from start to end=[start:] <br>
3.steps=[::step]

```python
print(arr1[1:5])
print(arr2[1,1:])
print(arr3[0,1,1:])
```

    it iterate array and is also can change data type for changinginng data type we have to provide np.nditen(var,flag='buffered',op_dtype='f')
### <br>2.np.enumurate():<br>
