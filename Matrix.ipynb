# Matrix
it perform all matrix operation

```python
mat=np.matrix([[1,2,3],[1,2,3]])
print(mat)
print(type(mat))
```

```python
mat2=np.matrix([[1,2,3],[1,2,3]])
print('addition: \n',mat+mat2)
print('substraction: \n',mat-mat2)
```

for multiplication in matrix we use .dot() for multiply

```python
mat3=np.matrix([[1,2],[1,2]])
print('multiplication: \n',mat3.dot(mat3))
```

## functions in matrix

### 1.Transpose: np.transpose()  (shortcut:var.T)
### 2.Swapaxes: np.swapaxes(var,change,change)
### 3.Inverse: np.linalg.inv()
### 4.Power: np.linalg.matrix_power(var,n)
### 5.Determinate: np.linalg.det()

```python
mat4=np.matrix([[1,2],[3,4]])
print('1.Transpose:\n', np.transpose(mat))
print('2.Swapaxes:\n', np.swapaxes(mat4,1,0))
print('3.Inverse: \n',np.linalg.inv(mat4))
print('4.Power: \n',np.linalg.matrix_power(mat4,2))
print('5.Determinate: \n',np.linalg.det(mat4))
```
